```python
import pandas as pd
```

```python
df3 = pd.read_excel('KYU_Score_Final.xlsx')
```

```python
df3
```

```python
def get_email_type(email):
    personal_domains = ['gmail.com', 'yahoo.com', 'hotmail.com']
    domain = email.split('@')[-1].lower()
    return 'Personal' if any(p in domain for p in personal_domains) else 'Organisational'

df3['Email_Type'] = df3['Email'].apply(get_email_type)
```

```python
df3
```

```python
#------2: Target Encoding-------
kyu_map = {'Low': 0, 'Moderate': 1, 'High': 2}
df3['KYU_Score_Encoded'] = df3['KYU Score'].map(kyu_map)
```

```python
df3.KYU_Score_Encoded.value_counts()
```

```python
#-------3: Feature Selection------
C = df3[['Email_Type', 'Domain', 'Purpose']]
d = df3['KYU_Score_Encoded']
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
```

```python
#-------4: One-hot encode categorical variables------
categorical_features = ['Email_Type', 'Domain', 'Purpose']
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ]
)
```

```python
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#------5: Train/test split------
C_train, C_test, d_train, d_test = train_test_split(C, d, test_size=0.2, random_state=42)
```

```python
# 6. Apply the one-hot encoding 
C_train_processed = preprocessor.fit_transform(C_train)
C_test_processed = preprocessor.transform(C_test)
```

```python
#-----6: Defining Models-----
RF = RandomForestClassifier()
DT = DecisionTreeClassifier()
LR = LogisticRegression()
```

```python
RF.fit(C_train_processed,d_train)
```

```python
DT.fit(C_train_processed,d_train)
```

```python
LR.fit(C_train_processed,d_train)
```

```python
d_pred = RF.predict(C_test_processed)
print(accuracy_score(d_test,d_pred))
```

```python
print(classification_report(d_test,d_pred))
```

```python
d_pred_2 = DT.predict(C_test_processed)
print(accuracy_score(d_test,d_pred_2))
```

```python
print(classification_report(d_test,d_pred_2))
```

```python
d_pred_3 = DT.predict(C_test_processed)
print(accuracy_score(d_test,d_pred_3))
```
